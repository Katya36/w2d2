```
import requests
import json
```

```
url = 'http://b6a442da9b9d.ngrok.io'
```

```
features = [10.0, 0.5, 0.47, 2.0, 0.171, 27.0, 66.0, 0.99820, 3.17, 0.76, 10.8]
```

```
req = requests.post(url,
                headers = {'Content-Type': 'application/json'},
                data = json.dumps({'features': features})
               )
```

```
req.json()['prediction']
```
